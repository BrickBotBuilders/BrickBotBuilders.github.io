```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
from numpy import pi
```

```python
x = np.arange(-2*pi,2*pi+pi/8,pi/8)
y = np.sin(x)
plt.ion()   # Turn interactive mode on
# Interactive = the plot gets updated as you go along
# For non-interactive, the plot doesn’t show up until you’ve finished everything
# to turn interactive mode off, enter the following command
# plt.ioff() 
plt.plot(x, y)
```

```python
# Change the line width:
plt.plot(x, y, linewidth=3)
```

```python
# to change the line style
plt.plot(x, y, linewidth=2, linestyle='dashed')
```

```python
# to change the line color
plt.plot(x, y, linewidth=1, linestyle='dashed', color='r')
```

```python
# to add point markers:
plt.plot(x, y, linewidth=1, linestyle='dashed', color='k', marker='d')
```

```python
# another marker
plt.plot(x, y, linewidth=1, linestyle='dashed', color='b', marker='*')
```

```python
# yet another marker
plt.plot(x, y, linewidth=3, linestyle='none', color='y', marker='s')
```

```python
# You can use LaTeX symbols as markers
plt.plot(x, y, linewidth=1, linestyle='none', color='m', marker='$\\beta$')
```

```python
# you can change the marker size
plt.plot(x, y, linewidth=1, linestyle='none', color='m', marker='$\\beta$', markersize = 10)
```

```python
# assume our y values have some error, shown by yerrrors. this error is random and let's say <= 0.4
yerrors = np.random.uniform(0,0.2,len(y))
```

```python
plt.errorbar(x, y, yerr=yerrors)
```

```python
# Creating error bars without lines:
plt.errorbar(x, y, yerr=yerrors, fmt='*')
```

Polar Plots and Alpha

```python
import numpy as np
import matplotlib.pyplot as plt

# Cluster data
a = np.random.uniform(0, 2*np.pi, size=10000)
r = np.random.exponential(size=10000)

# Make a polar plot.
plt.polar(a, r, linestyle='none', marker='o')
plt.show()
```

```python
# Alpha controls transparency
# 0 = transparent
# 1 = opaque

# Make markers transparent.
plt.clf()
plt.polar(a, r, linestyle='none', marker='o', alpha=0.2)
plt.show()
```

### HISTOGRAMS

```python
# Gaussian random data
y = np.random.normal(size=1000)

# Default histogram (10 bins)
plt.clf()
plt.hist(y)
```

```python
# Change the number of bins.
plt.clf()
plt.hist(y, bins=20)
```

```python
# Change the bin edges.
plt.clf()
edges=[-3, -2, -1, 0, 2]
plt.hist(y, bins=edges)
```

```python
# Just show the lines.
plt.clf()
plt.hist(y, histtype='step')
```

## SUBPLOTS

```python
# 3x1 grid of plots.
plt.clf()
x = np.linspace(-1, 1, 100)
y1 = x
y2 = x**2
y3 = x**3

# Make the plots one at a time.
ax1 = plt.subplot(3, 1, 1)
ax1.plot(x, y1)
ax2 = plt.subplot(3, 1, 2)
ax2.plot(x, y2)
ax3 = plt.subplot(3, 1, 3)
ax3.plot(x, y3)
```

```python
# Arbitrary subplot placement.
plt.clf()

# Make the plots one at a time.
ax1 = plt.axes([0.1, 0.1, 0.8, 0.8])
ax1.plot(x, y1)
ax2 = plt.axes([0.75, 0.75, 0.2, 0.2])
ax2.plot(x, y2)
ax3 = plt.axes([0.05, 0.75, 0.2, 0.2])
ax3.plot(x, y3)
```

```python
# EQUAL AXES 
import matplotlib.pyplot as plt
import numpy as np

# Let's create a circle of radius 3.
an = np.linspace(0, 2 * np.pi, 100)
x = 3 * np.cos(an)
y = 3 * np.sin(an)

fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x, y)
axs[0, 0].set_title('not equal, looks like ellipse', fontsize=10)

axs[0, 1].plot(x, y)
axs[0, 1].axis('equal')
axs[0, 1].set_title('equal, looks like circle', fontsize=10)

axs[1, 0].plot(x, y)
axs[1, 0].axis('equal')
axs[1, 0].axis([-3, 3, -3, 3])
axs[1, 0].set_title('still a circle, even after changing limits', fontsize=10)

axs[1, 1].plot(x, y)
axs[1, 1].set_aspect('equal', 'box')
axs[1, 1].set_title('still a circle, auto-adjusted data limits', fontsize=10)

fig.tight_layout()

plt.show()
```

##  TEXT ANOTATIONS
